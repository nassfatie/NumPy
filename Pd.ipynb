**Pandas DataFrames**
A Dataframe is a rectangular table of data and contains an ordered, named collection of columns each with different value type such as strings, numbers, boolean etc.

```python
data = {
    "Names":["Fania","Shania","Tahiya","Shuraim","Sameah","Hairah"],
    "Class": ["P.6","Baby","P.7","P.3","P.2","P.1"],
    "Age" : [12, 2, 13, 9, 7, 6]
}
Frame = pd.DataFrame(data)
print(Frame)
```

```python
print(type(data))
```

```python
new = pd.DataFrame(Frame,columns=["Names", "Age", "Class", "Address"])
print(new)
```
